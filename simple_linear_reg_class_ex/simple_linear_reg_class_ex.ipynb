```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = [5,7,8,7,2,17,9,5,4,3]
y = [99,85,88,112,93,111,78,75,88,99]

plt.scatter(x,y)
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(x,y)
#y = mx + c

def myfunc(x):
  return slope*x + intercept

mymodel = list(map(myfunc, x))
plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show()
print(r) # coefficient of correlation or area between pedicted and actual value


```
